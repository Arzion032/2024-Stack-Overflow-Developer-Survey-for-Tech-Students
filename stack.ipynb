### 4.2. How have salaries varied across tech jobs?

```python
from filter import cleaners
#Called the cleaneres function to transform the data
emp = cleaners(df, "Employment",dash=False)

#Dropped all rows where "ConvertedCompYearly" is null
new_emp = emp.dropna(subset="ConvertedCompYearly")

# Provide a list of column names where rows will be dropped if any of the specified columns contain the value 1.
# This is to filter out rows that indicate the respondent is not employed or is a student, to focus on those who are earning a salary.

columns_to_check_for_removal = [
    "Student, full-time", 
    "Not employed, but looking for work", 
    "Not employed, and not looking for work", 
    "Student, part-time", 
    "I prefer not to say"
]

# Filter the dataframe to include only individuals who are currently working
# This includes:
# - Full-time employees
# - Independent contractors, freelancers, or self-employed individuals
# - Part-time employees
# - Working students
filtered_emp = new_emp[((new_emp['Employed, full-time'] == 1) | (new_emp['Independent contractor, freelancer, or self-employed']==1) | (new_emp['Employed, part-time']==1)) & (new_emp["Not employed, but looking for work"].isnull()) & (new_emp['Not employed, and not looking for work'].isnull()) & (new_emp['Retired'].isnull()) & (new_emp['I prefer not to say'].isnull())]

# Remove the employment type columns
q2 = filtered_emp.loc[:,["DevType","ConvertedCompYearly","Country"]]

# ----------------------------
# Filtering Out Extreme Low Salaries
# ----------------------------
# Reference: Information on global minimum wages, including the Philippines,
# can be found at https://seasia.co/infographic/countries-with-the-lowest-minimum-wages-2024.
# The Philippines is among the top 5 countries with the lowest minimum wages according to this source.
# Therefore, we use the minimum wage in PHP to set a threshold for filtering out abnormally low salaries,

# Minimum monthly wage in PHP
monthly_salary_php = 128  

# Convert monthly wage to annual threshold
threshold = monthly_salary_php * 12 

# Filter out data points where the annual converted salary is below the defined threshold
q2 = q2[q2["ConvertedCompYearly"] >= threshold] 

# Reason for filtering:
# The dataset contained some extremely low salary values (e.g., $1 USD annually),
# which are likely data entry errors or anomalies. Including these values could
# distort the overall analysis by compressing the scale of the graph and making
# it difficult to interpret the more relevant salary ranges for tech job titles.
# By removing these extreme low values, we focus on salary ranges that are more
# representative of typical compensation, providing a clearer view of central trends.
#
# High salary values are retained because they represent valid, albeit rare,
# high-end salaries. These values can provide insights into the top end of the
# salary range and are less likely to distort the overall analysis compared to
# extremely low values.
#
# To better represent the central tendency of salaries and avoid distortion by
# extreme values, we use the median in our analysis. The median is less affected
# by outliers and provides a more accurate reflection of the typical salary
# range within the dataset.
#---------------------------------------------------------------------------------
# Group by 'DevType' and calculate median, min, max, and count for 'ConvertedCompYearly'
# The result is a summary DataFrame showing these statistics for each job type.

q2 = q2.groupby("DevType")["ConvertedCompYearly"].agg(['median','min','max','count']).reset_index()
```

### 4.3 Where do developers most commonly seek coding resources (e.g., books, online tutorials, technical documentation), and how do these preferences differ based on their years of coding experience?

```python
q3 = cleaners(df, "LearnCodeOnline",dash=False)
q3 = q3.drop(columns=["ConvertedCompYearly","DevType","Country","ResponseId"])
q3 = q3.dropna(subset="YearsCode")
```

```python
def categorize_experience(years):
    if years == 'Less than 1 year':
        return 'Beginner'
    elif years == 'More than 50 years':
        return 'Expert'
    try:
        years = int(years)
        if years < 6:
            return 'Beginner'
        elif 6 <= years <= 10:
            return 'Intermediate'
        elif 11 <= years <= 20:
            return 'Advanced'
        elif years > 20:
            return 'Expert'
    except ValueError:
        return 'Unknown'

q3["Coding Experience"] = q3["YearsCode"].apply(categorize_experience)
graph = q3.drop(columns="YearsCode")
```

```python
graph = graph.groupby('Coding Experience').sum().reset_index()
graph = graph.drop(columns="Other")
#print(graph,'\n')
graph = graph.set_index('Coding Experience').transpose()
#print(graph,'\n')
graph.index.name = 'Resource'
#print(graph,'\n')
graph['Sum'] = graph[['Advanced', 'Beginner', 'Expert', 'Intermediate']].sum(axis=1)
graph_reset = graph.reset_index()
#print(graph,'\n')
graph_reset.rename(columns={'index': 'Resource'}, inplace=True)
graph = graph.sort_values(by="Sum",ascending = False)
#print(graph,'\n')
```

### 4.4 What factors have the most significant impact on job satisfaction among developers?

```python
# Choose 'YearsCode', 'Salary', and 'RemoteWork' to analyze their relationship with job satisfaction ('JobSat').
# - 'YearsCode' and 'Salary' are continuous variables that might affect job satisfaction.
# - 'RemoteWork' is a categorical variable with an ordinal order (Remote > Hybrid > In-person) that could influence job satisfaction.

# Using Spearman's rank correlation for the following reasons:
# 1. It measures monotonic relationships, which suits both continuous variables (YearsCode, Salary) and ordinal data (RemoteWork).
# 2. It does not assume a linear relationship, making it appropriate for the ordinal nature of RemoteWork and the rating scale of JobSat.

# This method helps to understand the strength and direction of the association between each factor and job satisfaction.
```

```python
# Filter the dataframe to include only individuals who are currently working
# This includes:
# - Full-time employees
# - Independent contractors, freelancers, or self-employed individuals
# - Part-time employees
# - Working students

#Transform the data for filtering 
q4 = cleaners(df,"Employment", dash = False)

#Filter the data
q4 = q4[((q4['Employed, full-time'] == 1) | (q4['Independent contractor, freelancer, or self-employed']==1) | (q4['Employed, part-time']==1)) & (q4["Not employed, but looking for work"].isnull()) & (q4['Not employed, and not looking for work'].isnull()) & (q4['Retired'].isnull()) & (q4['I prefer not to say'].isnull())]

#Now that the data is filtered to only include those who are working
#We can now remove the employment columns
q4 = result = pd.merge(q4, df, on='ResponseId', how='left', suffixes=('_left', '_right'))
q4 = q4.loc[:,['ResponseId','JobSat','ConvertedCompYearly_left','YearsCode_left','RemoteWork']]
q4 = q4.rename(columns={
    'ConvertedCompYearly_left': 'ConvertedCompYearly',
    'YearsCode_left': 'YearsCode'})
#Null values handled
q4 = q4.dropna()
```

```python
# Years_code transformation
years_code_map = {'More than 50 years' : 51, 'Less than 1 year' : 0.5}

def transform_years_code(years):
    if years in years_code_map:
        return years_code_map[years]
    else:
        try:
            return float(years)
        except ValueError:
            return None
        
# Remote Work Transformation: The explanation below        
remote_work_map = {'In-person': 0,'Hybrid (some remote, some in-person)': 0.5, 'Remote': 1}

def transform_remote_work(work):
    if work in remote_work_map:
        return remote_work_map[work]
    else: 
        return work
# Apply each function to q4 data frame
apply_func = {'YearsCode' : transform_years_code, 'RemoteWork': transform_remote_work}

for column, func in apply_func.items():
    q4[column] = q4[column].apply(func)
    
# Now we will remove all the outlier rows from the 'ConvertedCompYearly' (Salary) column using the IQR method
# We first need to calculate Q1 (25th percentile) and Q3 (75th percentile) for Salary
Q1 = q4['ConvertedCompYearly'].quantile(0.25)
Q3 = q4['ConvertedCompYearly'].quantile(0.75)

# Then use the 1st and 3rd quartile to calculate the IQR
IQR = Q3 - Q1

# Determine the bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data, removing outliers
q4 = q4[(q4['ConvertedCompYearly'] >= lower_bound) & (q4['ConvertedCompYearly'] <= upper_bound)]
```

```python
"""Analysis of Remote Work and Job Satisfaction

Methodology:
We employed Spearman's rank correlation to explore the relationship between remote work arrangements and job satisfaction. Remote work categories were transformed into an ordinal scale (1: In-person, 2: Hybrid, 3: Remote) based on the assumption of increasing "remoteness".

Results:
The Spearman correlation coefficient was [value] with a p-value of [value]. [Interpret based on your actual results]

Limitations:
This analysis assumes an ordinal relationship between remote work categories, which may not fully capture the complexity of work arrangements. The method doesn't account for potential confounding variables or provide information about specific category differences.

Interpretation:
[Your cautious interpretation based on results]

Future Directions:
We plan to conduct a one-way ANOVA in the future, which may provide more robust insights into the relationship between remote work categories and job satisfaction. Additionally, we aim to control for other variables that might influence job satisfaction in future analyses."""

spearman_years, _ = stats.spearmanr(q4['YearsCode'], q4['JobSat'])
spearman_salary, _ = stats.spearmanr(q4['ConvertedCompYearly'], q4['JobSat'])
spearman_remote, _ = stats.spearmanr(q4['RemoteWork'], q4['JobSat'])

# Set up the plot style
sns.set(style="whitegrid")

# 1. Scatter plot: YearsCode vs JobSat
plt.figure(figsize=(12.5, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(x=q4['YearsCode'], y=q4['JobSat'])
plt.title('YearsCode vs JobSat')
plt.xlabel('Years of Coding')
plt.ylabel('Job Satisfaction')

# 2. Scatter plot: Salary vs JobSat
plt.subplot(1, 3, 2)
sns.scatterplot(x=q4['ConvertedCompYearly'], y=q4['JobSat'])
plt.title('Salary vs JobSat')
plt.xlabel('Salary (Yearly)')
plt.ylabel('Job Satisfaction')

# 3. Box plot: RemoteWork vs JobSat
remote_work_labels = {0: 'In-person', 0.5: 'Hybrid', 1: 'Remote'}
q4['RemoteWorkLabels'] = q4['RemoteWork'].map(remote_work_labels)

# 3. Box plot: RemoteWork (categorized) vs JobSat
plt.subplot(1, 3, 3)
sns.boxplot(x=q4['RemoteWorkLabels'], y=q4['JobSat'])  # Use the mapped 'RemoteWorkLabels'
plt.title('RemoteWork vs JobSat')
plt.xlabel('Remote Work Type')  # Now this will show 'In-person', 'Hybrid', and 'Remote'
plt.ylabel('Job Satisfaction')

plt.tight_layout()
plt.show()

print('Spearman correlation between YearsCode and JobSat    :', spearman_years)
print('Spearman correlation between Salary and JobSat       :', spearman_salary)
print('Spearman correlation between RemoteWork and JobSat   :', spearman_remote)
```

**Findings:**
